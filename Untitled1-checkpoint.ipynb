```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.optim.lr_scheduler import StepLR, CosineAnnealingLR, ReduceLROnPlateau
import time
from thop import profile
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
from tabulate import tabulate
```

```python

torch.manual_seed(42)
np.random.seed(42)


device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# 数据增强和加载
transform_train = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
])

transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
])

# 加载CIFAR-10数据集
train_dataset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_train)
test_dataset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_test)

batch_size = 128
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2)
```

```python
# 定义中等规模的CNN模型
class MediumCNN(nn.Module):
    def __init__(self):
        super(MediumCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 64, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(64)
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(128)
        self.conv3 = nn.Conv2d(128, 256, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(256)
        self.conv4 = nn.Conv2d(256, 512, kernel_size=3, padding=1)
        self.bn4 = nn.BatchNorm2d(512)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(512 * 2 * 2, 1024)
        self.fc2 = nn.Linear(1024, 10)
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        x = self.pool(torch.relu(self.bn1(self.conv1(x))))
        x = self.pool(torch.relu(self.bn2(self.conv2(x))))
        x = self.pool(torch.relu(self.bn3(self.conv3(x))))
        x = self.pool(torch.relu(self.bn4(self.conv4(x))))
        x = x.view(-1, 512 * 2 * 2)
        x = self.dropout(torch.relu(self.fc1(x)))
        x = self.fc2(x)
        return x
```

```python
# 训练和测试函数
def train(model, device, train_loader, optimizer, criterion, epoch):
    model.train()
    train_loss = 0
    correct = 0
    total = 0
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)
        optimizer.zero_grad()
        output = model(data)
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()

        train_loss += loss.item()
        _, predicted = output.max(1)
        total += target.size(0)
        correct += predicted.eq(target).sum().item()

    acc = 100. * correct / total
    avg_loss = train_loss / len(train_loader)
    return avg_loss, acc
```

```python
def test(model, device, test_loader, criterion):
    model.eval()
    test_loss = 0
    correct = 0
    total = 0
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            loss = criterion(output, target)

            test_loss += loss.item()
            _, predicted = output.max(1)
            total += target.size(0)
            correct += predicted.eq(target).sum().item()

    acc = 100. * correct / total
    avg_loss = test_loss / len(test_loader)
    return avg_loss, acc
```

```python
# 计算模型FLOPs和参数数量
def calculate_flops_params(model):
    input = torch.randn(1, 3, 32, 32).to(device)
    flops, params = profile(model, inputs=(input,))
    return flops, params

# 实验配置
epochs = 20
criterion = nn.CrossEntropyLoss()

# 定义不同的实验配置
experiments = {
    'SGD_lr0.1': {
        'optimizer': optim.SGD,
        'optim_args': {'lr': 0.1, 'momentum': 0.9, 'weight_decay': 5e-4},
        'scheduler': StepLR,
        'sched_args': {'step_size': 30, 'gamma': 0.1},
        'reg': None
    },
    'SGD_lr0.1_wd0': {
        'optimizer': optim.SGD,
        'optim_args': {'lr': 0.1, 'momentum': 0.9, 'weight_decay': 0},
        'scheduler': StepLR,
        'sched_args': {'step_size': 30, 'gamma': 0.1},
        'reg': None
    },
    'Adam_lr0.001': {
        'optimizer': optim.Adam,
        'optim_args': {'lr': 0.001, 'weight_decay': 5e-4},
        'scheduler': None,
        'sched_args': None,
        'reg': None
    },
    'RMSprop_lr0.001': {
        'optimizer': optim.RMSprop,
        'optim_args': {'lr': 0.001, 'weight_decay': 5e-4},
        'scheduler': ReduceLROnPlateau,
        'sched_args': {'mode': 'min', 'factor': 0.1, 'patience': 5},
        'reg': None
    },
    'Adam_Cosine': {
        'optimizer': optim.Adam,
        'optim_args': {'lr': 0.001, 'weight_decay': 5e-4},
        'scheduler': CosineAnnealingLR,
        'sched_args': {'T_max': epochs},
        'reg': None
    }
}

# 运行实验
results = defaultdict(dict)
```

```python
for exp_name, config in experiments.items():
    print(f"\nRunning experiment: {exp_name}")

    # 初始化模型
    model = MediumCNN().to(device)
    
    # 初始化优化器
    optimizer = config['optimizer'](model.parameters(), **config['optim_args'])

    # 初始化学习率调度器
    scheduler = None
    if config['scheduler'] is not None:
        scheduler = config['scheduler'](optimizer, **config['sched_args'])

    # 训练前计算FLOPs和参数数量
    flops, params = calculate_flops_params(model)
    results[exp_name]['flops'] = flops
    results[exp_name]['params'] = params

    # 训练模型
    train_losses = []
    train_accs = []
    test_losses = []
    test_accs = []
    start_time = time.time()

    for epoch in range(1, epochs + 1):
        train_loss, train_acc = train(model, device, train_loader, optimizer, criterion, epoch)
        test_loss, test_acc = test(model, device, test_loader, criterion)

        train_losses.append(train_loss)
        train_accs.append(train_acc)
        test_losses.append(test_loss)
        test_accs.append(test_acc)

        # 更新学习率
        if scheduler is not None:
            if isinstance(scheduler, ReduceLROnPlateau):
                scheduler.step(test_loss)
            else:
                scheduler.step()

        print(f'Epoch {epoch}: Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%, '
              f'Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.2f}%')

    # 记录结果
    end_time = time.time()
    results[exp_name]['train_time'] = end_time - start_time
    results[exp_name]['final_train_acc'] = train_accs[-1]
    results[exp_name]['final_test_acc'] = test_accs[-1]
    results[exp_name]['best_test_acc'] = max(test_accs)
    results[exp_name]['train_losses'] = train_losses
    results[exp_name]['test_losses'] = test_losses
    results[exp_name]['train_accs'] = train_accs
    results[exp_name]['test_accs'] = test_accs

    # 保存模型
    torch.save(model.state_dict(), f'model_{exp_name}.pth')
```

```python
# 创建多子图对比可视化
plt.figure(figsize=(18, 12))

# 子图1：测试准确率对比（原有代码）
plt.subplot(2, 2, 1)
for exp_name in experiments.keys():
    plt.plot(results[exp_name]['test_accs'], label=exp_name, linewidth=2)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Test Accuracy (%)', fontsize=12)
plt.title('Test Accuracy Comparison', fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(0, 100)  # 固定Y轴范围便于对比

# 子图2：训练损失对比（新增）
plt.subplot(2, 2, 2)
for exp_name in experiments.keys():
    plt.plot(results[exp_name]['train_losses'], '--', label=exp_name, linewidth=2)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Training Loss', fontsize=12)
plt.title('Training Loss Comparison', fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.yscale('log')  # 对数坐标更易观察变化

# 子图3：测试损失对比（新增）
plt.subplot(2, 2, 3)
for exp_name in experiments.keys():
    plt.plot(results[exp_name]['test_losses'], '-', label=exp_name, linewidth=2)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Test Loss', fontsize=12)
plt.title('Test Loss Comparison', fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.yscale('log')

# 子图4：最终指标对比（新增）
plt.subplot(2, 2, 4)
final_metrics = ['final_train_acc', 'final_test_acc', 'best_test_acc']
x = np.arange(len(experiments))
width = 0.25
for i, metric in enumerate(final_metrics):
    values = [results[exp_name][metric] for exp_name in experiments.keys()]
    plt.bar(x + i*width, values, width, label=metric.replace('_', ' ').title())
plt.xlabel('Experiments', fontsize=12)
plt.ylabel('Accuracy (%)', fontsize=12)
plt.title('Final Metrics Comparison', fontsize=14)
plt.xticks(x + width, experiments.keys(), rotation=45, ha='right')
plt.legend(fontsize=10)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, 100)

plt.tight_layout()
plt.savefig('full_metrics_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python


# 准备表格数据
table_data = []
headers = ["Experiment", "FLOPs(M)", "Params(M)", "Train Time(s)", 
           "Final Train Acc(%)", "Final Test Acc(%)", "Best Test Acc(%)"]

for exp_name in experiments.keys():
    row = [
        exp_name,
        f"{results[exp_name]['flops'] / 1e6:.2f}",
        f"{results[exp_name]['params'] / 1e6:.2f}",
        f"{results[exp_name]['train_time']:.2f}",
        f"{results[exp_name]['final_train_acc']:.2f}",
        f"{results[exp_name]['final_test_acc']:.2f}",
        f"{results[exp_name]['best_test_acc']:.2f}"
    ]
    table_data.append(row)

# 打印表格
print("\nFinal Results:")
print(tabulate(table_data, headers=headers, tablefmt="grid", floatfmt=".2f"))
```
